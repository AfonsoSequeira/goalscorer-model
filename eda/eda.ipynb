```python
import pandas as pd
import os
```

```python
data_code_dict = {
    '9':'Premier League',
    '11':'Serie A',
    '12':'Bundesliga'
}

def read_data(fileDir, league_name, season):
    data = pd.read_csv(fileDir)
    data["league"] = league_name
    data["season"] = season
    return data
```

```python
base_dir = "C:/MyDevelopment/Goalscorers/data/"
data = []

for file in os.listdir(base_dir):
    if "summary" in file:
        fh = file.split("-")
        season = "-".join(fh[0:2])
        league = data_code_dict[fh[3]]
        season_data = pd.read_csv(base_dir+file)
        season_data["league_name"] = league
        season_data["season"] = season
        data.append(season_data)
        
data = pd.concat(data, axis=0)
```

```python
data['datetime'] = pd.to_datetime(data['datetime'], unit='s')
```

```python
data = data.query("league_name == 'Premier League'")
```
